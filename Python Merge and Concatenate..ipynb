Etract the two data we are joining together

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\THINKPAD T480s\OneDrive\Desktop\ife\dataframe 1.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\THINKPAD T480s\OneDrive\Desktop\ife\dataframe 2.csv")
df2
```

merge the ones similar to each other together

```python
df1.merge(df2)
```

inner join data and split df1 & df2 in same line 

```python
df1.merge(df2, how = 'inner', on = 'ID')
```

Outer Join df1 and df2 together 

```python
df1.merge(df2, how = 'outer')
```

left Join df1 and df2 together 

```python
df1.merge(df2, how = 'left')
```

Right Join df1 and df2 together 

```python
df1.merge(df2, how = 'right')
```

Cross left Join df1 and df2 together 

```python
df1.merge(df2, how = 'cross')
```

Concatenate Join (The two table will be on top each other

```python
pd.concat([df1,df2])
```

```python
Concatenate outer Join (The two table will be by the side of each other)
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```
